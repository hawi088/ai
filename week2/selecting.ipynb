```python
students = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack'],
    'Math_Score': [85, 72, 93, 88, 76, 95, 82, 79, 91, 84],
    'English_Score': [88, 70, 85, 92, 78, 88, 85, 81, 89, 86],
    'Science_Score': [90, 75, 88, 94, 80, 92, 86, 77, 93, 88]
})
```

```python
#Get name and Math_Score as a dataframe
print(students[['Name','Math_Score']])
```

```python
print(students[students['Math_Score'] > 85])
```

```python
print(students[(students['Math_Score'] > 85) & (students['English_Score'] >80)])
```

```python
students['Average'] = students[['Math_Score','English_Score','Science_Score']].mean(axis=1)
print(students.head())
```

```python
def get_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 80:
        return 'B'
    elif avg >= 70:
        return 'C'
    else:
        return 'D'
```

```python
get_grade(80)
```

```python
students['Grade'] = students['Average'].apply(get_grade)
print(students.head())
```
